# Update Documents and collections

```python
from conect import db
from pymongo import UpdateOne

modificar = db.libros

# usando UpdateOne para agregar 10 libros más
res = modificar.update_one(
    {'Titulo': 'La ciudad Verde'},   # filtro
    {'$inc': {'Stock': 10}}          # actualización
)

print('Cambios',res.matched_count, 'Total',res.modified_count)
```
